```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import AxesGrid
plt.rcParams["pdf.use14corefonts"] = True


def plot_activations_new(vals, rc, labels, x_labels, normalize=False, filename=None):

    font = {'family': 'serif',
            'color':  'black',
            'weight': 'normal',
            'size': 18,
            }
    fig = plt.figure()
    fig.set_size_inches(10, 5)

    grid = AxesGrid(fig, 111,
                    nrows_ncols=rc,
                    axes_pad=0.0,
                    share_all=True,
                    label_mode="L",
                    cbar_location="right",
                    cbar_mode="single",
                    cbar_size="1%"
                   )
    
    for i, (val, ax) in enumerate(zip(vals,grid)):
        ax.tick_params(labelsize=10)
        #ax.set_xlabel('Activation position', fontdict=font, labelpad=15)
        ax.set_xlabel(x_labels[i % 2], fontdict=font, labelpad=15)
        ax.set_ylabel(labels[i], fontdict=font)
        if normalize:
            im = ax.imshow(val, cmap='gist_heat_r',vmin=0)#, vmax=20)
        else:
            im = ax.imshow(val, cmap='gist_heat_r')

    grid.cbar_axes[0].colorbar(im)

    for cax in grid.cbar_axes:
        cax.toggle_label(False)
        
    if filename:
        plt.savefig("figures/" + filename + ".pdf")
    plt.show()
```

```python
# path to the h5 dataset file - single-proxy traces
h5_path_S = "data/tc_S_flowlen-100_neg-49_cv-5.h5"
config_yml_S = "config-8k-S.yml"

# path to the h5 dataset file - multi-proxy traces
h5_path_M = "data/tc_M_flowlen-100_neg-49_cv-5.h5"
config_yml_M = "config-8k-M.yml"
```

```python
res={}

folds = range(5)
runs = [10]

for fold in folds:
    
    with h5py.File(h5_path_S, 'r') as f:
        test_indices = list(f['indices'][f'test{fold}'])
        x = list(f['data']['x'])
        y = list(f['data']['y'])

    # Only correlated
    s_test_x = [x[index] for index in test_indices if y[index] == 1]
    s_test_y = [1] * len(s_test_x)

    s_test_x = tf.constant(np.array(s_test_x).astype('float32'))
    s_test_y = tf.constant(np.array(s_test_y).astype('float32'))
#     print("Test pos X:", test_x.shape)

    # Not correlated
    s_neg_test_x = [x[index] for index in test_indices if y[index] != 1]
    s_neg_test_y = [0] * len(s_neg_test_x)

    s_neg_test_x = tf.constant(np.array(s_neg_test_x).astype('float32'))
    s_neg_test_y = tf.constant(np.array(s_neg_test_y).astype('float32'))
#     print("Test neg X:", neg_test_x.shape)

    s_neg_test_x = s_neg_test_x[:s_test_x.shape[0]]
#     print("Test neg X:", neg_test_x.shape)
    
    with h5py.File(h5_path_m, 'r') as f:
        test_indices = list(f['indices'][f'test{fold}'])
        x = list(f['data']['x'])
        y = list(f['data']['y'])

    # Only correlated
    m_test_x = [x[index] for index in test_indices if y[index] == 1]
    m_test_y = [1] * len(m_test_x)

    m_test_x = tf.constant(np.array(m_test_x).astype('float32'))
    m_test_y = tf.constant(np.array(m_test_y).astype('float32'))
#     print("Test pos X:", test_x.shape)

    # Not correlated
    m_neg_test_x = [x[index] for index in test_indices if y[index] != 1]
    m_neg_test_y = [0] * len(m_neg_test_x)

    m_neg_test_x = tf.constant(np.array(m_neg_test_x).astype('float32'))
    m_neg_test_y = tf.constant(np.array(m_neg_test_y).astype('float32'))
#     print("Test neg X:", neg_test_x.shape)

    m_neg_test_x = m_neg_test_x[:m_test_x.shape[0]]
#     print("Test neg X:", neg_test_x.shape)

    for seed in runs:
        t_m  = Server(cf="config_yml_M", fold=fold, pred_only=True, init_data=False, run=seed)
        t_s = Server(cf="config_yml_S", fold=fold, pred_only=True, init_data=False, run=seed)

#         with open("log_tc_M_neg-49_CV5_FOLDS/train/train_best_aux_tc_correct_M_fold{}-{}.pkl".format(fold,seed), "rb") as f:
#             t_m.model = pickle.load(f)

#         with open("log_tc_S_neg-49_CV5_FOLDS/train/train_best_aux_tc_correct_S_fold{}-{}.pkl".format(fold,seed), "rb") as f:
#             t_s.model = pickle.load(f)


        output_single = t_s.model.get_activations(s_test_x, name=t_s.model.layers[0].layers[0].name)
        output_single_neg = t_s.model.get_activations(s_neg_test_x[:1590], name=t_s.model.layers[0].layers[0].name)

#        print(output_single.shape)
        output_single_mean = np.mean(output_single, axis=0)

        Td = np.transpose(output_single_mean[0])
        Tu = np.transpose(output_single_mean[1])
        Sd = np.transpose(output_single_mean[2])
        Su = np.transpose(output_single_mean[3])
#        print(Td.shape, Sd.shape)

#        print(output_single_neg.shape)
        output_single_neg_mean = np.mean(output_single_neg, axis=0)

        Td_neg = np.transpose(output_single_neg_mean[0])
        Tu_neg = np.transpose(output_single_neg_mean[1])
        Sd_neg = np.transpose(output_single_neg_mean[2])
        Su_neg = np.transpose(output_single_neg_mean[3])
#         print(Td_neg.shape, Sd_neg.shape)

        single_act = {}
        single_act["pos"] = [Td,Tu,Sd,Su]
        single_act["neg"] = [Td_neg,Tu_neg,Sd_neg,Su_neg]

        output_multi = t_m.model.get_activations(m_test_x, name=t_m.model.layers[0].layers[0].name)
        output_multi_neg = t_m.model.get_activations(m_neg_test_x[:1590], name=t_m.model.layers[0].layers[0].name)

#         print(output_multi.shape)
        output_multi_mean = np.mean(output_multi, axis=0)

        Td = np.transpose(output_multi_mean[0])
        Tu = np.transpose(output_multi_mean[1])
        Sd = np.transpose(output_multi_mean[2])
        Su = np.transpose(output_multi_mean[3])
#         print(Td.shape, Sd.shape)

#         print(output_multi_neg.shape)
        output_multi_neg_mean = np.mean(output_multi_neg, axis=0)

        Td_neg = np.transpose(output_multi_neg_mean[0])
        Tu_neg = np.transpose(output_multi_neg_mean[1])
        Sd_neg = np.transpose(output_multi_neg_mean[2])
        Su_neg = np.transpose(output_multi_neg_mean[3])
#         print(Td_neg.shape, Sd_neg.shape)

        multi_act = {}
        multi_act["pos"] = [Td,Tu,Sd,Su]
        multi_act["neg"] = [Td_neg,Tu_neg,Sd_neg,Su_neg]


        print("Fold ", fold)
        print("SINGLE\tMULTI")
        plot_activations_new([abs(single_act["pos"][0]-single_act["neg"][0]), abs(multi_act["pos"][0]-multi_act["neg"][0]),
                              abs(single_act["pos"][1]-single_act["neg"][1]), abs(multi_act["pos"][1]-multi_act["neg"][1])],
                              labels = ["Td", "", "Tu", ""],
                              normalize = True,
                              rc=(2, 2))
                              #filename = f"new_activations_{seed}_{fold}")

        single=(sum(sum(abs(single_act["pos"][0]-single_act["neg"][0]))),sum(sum(abs(single_act["pos"][1]-single_act["neg"][1])))) # Tu
        multi=(sum(sum(abs(multi_act["pos"][0]-multi_act["neg"][0]))),sum(sum(abs(multi_act["pos"][1]-multi_act["neg"][1]))))   

        res[f"{seed}_{fold}"]={'single':single,'multi':multi}
```
